```python
file_conf = 'COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
conf = pd.read_csv(file_conf)
```

```python
conf.head()
```

```python
dates = [datetime.datetime.strptime(x, '%m/%d/%y') for x in conf.columns[4:]]
```

```python
ts = conf.set_index('Country/Region')
ts = ts.drop(['Province/State', 'Lat', 'Long'], axis=1).T
ts.index = dates
```

```python
last = ts.iloc[-1]
countries_sorted = last.sort_values(ascending=False).index
```

```python
ts[countries_sorted].plot()
```
